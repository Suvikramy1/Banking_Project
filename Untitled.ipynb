```python
from scipy.stats import norm
from scipy import stats
df_num.describe()
df_num['previous'].value_counts()
def fn_Validate_Norm(df_num):
    for i in ['age']:
        ks_test = stats.kstest(df_num[i],'norm')
        if(ks_test[1]>0.05):
            lower_boundary=df_num[i].mean() - 3*df_num[i].std()
            upper_boundary=df_num[i].mean() + 3*df_num[i].std()
        else:
            IQR = df_num[i].quantile(0.75) - df_num[i].quantile(0.25)
            lower_boundary= df_num[i].quantile(0.25) -3*IQR
            upper_boundary= df_num[i].quantile(0.75) +3*IQR
        df_num.loc[df_num[i]<lower_boundary,[i]]=lower_boundary
        df_num.loc[df_num[i]>upper_boundary,[i]]=upper_boundary

fn_Validate_Norm(df_num)
```

```python
learning_rate = 0.01
training_epochs = 25
batch_size = 100
display_step = 1
#df.drop(['marital','y_cat','duration'],axis=1,inplace=True)
df.drop(['y_cat'],axis=1,inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_X, y, test_size=0.3, random_state=0)

X_train.shape, X_test.shape

#df_X = tf.data.Dataset.from_tensors(X_train.values)
```

```python
X = tf.placeholder(tf.float32, [None,X_train.shape[1] ]) 
Y=tf.placeholder(tf.float32, [None, 2])
y_train.shape,y_test.shape
```

```python
# Create model
# Set model weights
W = tf.Variable(tf.zeros([X_train.shape[1],2]))
b = tf.Variable(tf.zeros([2]))
df_train.shape
```

```python
Y_hat = tf.nn.sigmoid(tf.add(tf.matmul(X, W), b))
cost = tf.nn.sigmoid_cross_entropy_with_logits(logits = Y_hat, labels = Y)  
# Softmaxtf.nn.sigmoid_cross_entropy_with_logits
```

```python
#cost = tf.nn.sigmoid_cross_entropy_with_logits(logits = Y_hat, labels = Y) 
# Minimize error using cross entropy
# Cross entropy
#cost = -tf.reduce_sum(Y*tf.log(activation)) 
# Gradient Descent
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost) 
```

```python
# Initializing the variables
init = tf.initialize_all_variables()
```

```python
#launch the session
with tf.Session() as ss:
    ss.run(init)
    avg_cost = 0.
    # Training cycle
    for epoch in range(training_epochs):
        avg_cost =0
        #total_batch = int(X_train.shape[0]/batch_size)
        # Loop over all batches
        batch_xs, batch_ys = X_train, y_train
        # Fit training using batch data
        ss.run(optimizer, feed_dict={X: batch_xs, Y: batch_ys})
        # Compute average loss
        avg_cost = avg_cost+ss.run(cost, feed_dict={X: batch_xs, Y: batch_ys})
        # Display logs per epoch step
        if epoch % display_step == 0:
            print("Epoch:", '%04d' % (epoch+1), "cost=", "{:.9f}".format(sum(sum(avg_cost))))

    print("Optimization Finished!")
    correct_prediction = tf.equal(tf.argmax(Y_hat, 1), 
                              tf.argmax(Y, 1)) 
    accuracy = tf.reduce_mean(tf.cast(correct_prediction,  
                                             tf.float32)) 
    print("\nAccuracy:", accuracy.eval({X:X_train,Y:y_train}), "%")
    print("\nAccuracy:", accuracy.eval({X:X_test,Y:y_test}), "%")
```
